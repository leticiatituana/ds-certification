## Dataframes ##

Los dataframes son la estructura de datos básica de Pandas. Es el equivalente a una tabla de una base de datos, con filas y columnas. Las columnas por lo general representan los diferentes campos que conforman una tabla, mientras que cada fila representa un registro de dicha tabla. Se puede cargar información de un archivo en un dataframe usando:

* pd.read_csv(<nombre_archivo>): Lee un archivo .csv o .txt o .tsv y lo carga en dataframe
* pd.read_excel(<nombre_archivo>): Lee un archivo .xls o .xlsx y lo carga en un dataframe

```python
# Para cargar un archivo.
df=pd.read_csv('Dataset Titanic.csv')
```

## Series ##

Una serie es un arreglo unidimensional y representan una columna individual (sea o no de un dataframe). Un dataframe está compuesto de un conjunto de series con un mismo índice

```python
# Seleccionar la serie (columna) de nombres
df["Age"].describe()
```

Muchas de las funciones de los dataframes pueden utilizarse para las series

```python
# Seleccionar las series de nombre y cabina
df[["Name","Cabin"]]
```

## Slicing de dataframes ##

Así como con otras estructuras de datos como las listas y los arreglos, los dataframes permiten realizar slicing para acceder a diferentes partes del mismo. Las principales funciones a utilizar son:

* df.loc[]: utiliza los identificadores de filas y columnas
* df.iloc[]: utiliza los índices numéricos de filas y columnas

```python
# Al visualizar un dataframe...
df.head()
```

```python
df_names=df.set_index("Name")
df_names
```

```python
df_names.columns
```

```python
#df.loc[:,:]
df_names.loc["Heikkinen, Miss. Laina":"Johnston, Miss. Catherine Helen "Carrie""]]
```

```python
#df.loc[:,:]
df_names.iloc[0:5,0:3]
```

```python
# La función "loc" puede recibir listas para realizar slicing
df_names["Moran, Mr James":,"Sex":"Embarked"]
```

## Mostrar filas por condición ##

```python
# Ver todos los pasajeros de tercera clase -> Pclass
df[df["Pclass"]==3]
```

```python
# Ver todos los pasajeros de mayores a 30 años -> Age
df[df["Age"]>30]
```

```python
# Ver todos los registros donde no tenga datos de cabina
df[df['Cabin'].isna()]
```

```python
# Ver todos los pasajeros de primera clase que hayan sobrevivido
df[(df["Pclass"]==1) & (df["Survived"]==1)]
```

```python
# Ver los registros de las cabinas tipo C
```

```python
df.loc[df["Cabin"].str.startswith("C")]
```

## Manejo de datos faltantes ##
* Sustituir valores vacíos
* Eliminar filas con valores vacíos

```python
# df.copy() -> Genera una copia del dataframe
df_copy=df.copy()
df_copy
```

```python
# df.fillna() -> Sustituye valores vacios
df_copy["Cabin"].fillna("No encontrada")
```

```python
# df.dropna() -> Eliminar filas con valores vacíos
df_copy=df.copy()
df_copy=df_copy.dropna(subset="Cabin")
df_copy
```
